# Explore Classify Fine Tuned Modles Results
Notebook for exporing the classify results for the fine tuned models.
The result are stored in `../output/evaluation_results/ classify_finetuned_*.csv`.  
The classify DSPy results contain standard ML metrics such as accuracy, precission, etc.  

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

----

## Load all classifaction results

```python
classify_files = glob.glob('../output/evaluation_results/classify_finetuned_*.csv')
len(classify_files)
```

```python
classify_df = pd.concat([pd.read_csv(file) for file in classify_files])
classify_df.head()
```

----

## Explore overall scores

looking at just f1_score, the non-agent large llms w/o dspy do best.

```python
overall_df = classify_df.query("property == 'all'").sort_values(by='f1_score', ascending=False)
overall_df.head(20)
```

---

## Expore rigidity  

```python
rigidity_df = classify_df.query("property == 'rigidity'").sort_values(by='f1_score', ascending=False)
rigidity_df[rigidity_df['class'] == 'all'].head(20)
```

```python
rigidity_df[rigidity_df['class'] == '+R'].sort_values(by='f1_score', ascending=False).head(20)
```

```python
rigidity_df[rigidity_df['class'] == '-R'].sort_values(by='f1_score', ascending=False).head(20)
```

```python
rigidity_df[rigidity_df['class'] == '~R'].sort_values(by='f1_score', ascending=False).head(20)
```

```python
rigidity_df['class'].unique()
```
