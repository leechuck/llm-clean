# Explore Classify DSPy Results
Notebook for exporing the classify results using the DSPy models.
The result are stored in `../output/evaluation_results/`.  
The classify DSPy results contain standard ML metrics such as accuracy, precission, etc.  

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

----

## Load all classifaction results

```python
classify_files = glob.glob('../output/evaluation_results/classify_*dspy_*.csv')
len(classify_files)
```

```python
classify_df = pd.concat([pd.read_csv(file) for file in classify_files])
classify_df.head()
```

----

## Explore overall scores

looking at just f1_score, the non-agent large llms w/o dspy do best.

```python
overall_df = classify_df.query("property == 'all'").sort_values(by='f1_score', ascending=False)
overall_df.head(20)
```

---

## Expore rigidity  
summary:  
* the gemini-crtic-no-intro had the best overall f1_score for all the rigidity classes (0.91)
* the gemini-crtic-no-intro, anthropic-corrected,  and had the best overall f1_score for +R (1.00)
* the anthropic and gemini critics

```python
rigidity_df = classify_df.query("property == 'rigidity'").sort_values(by='f1_score', ascending=False)
rigidity_df[rigidity_df['class'] == 'all'].head(20)
```

## Find best DSPy models from each of dataset of:
* overall scores
* dependence scores
* identity scores
* own identity scores
* unity scores
* dependence scores

Results in 6 models

```python
top_n = 1
top_models = list(
    set( pd.concat([
        overall_df.head(top_n),
        rigidity_df.head(top_n),
        identity_df.head(top_n),
        own_identity_df.head(top_n),
        unity_df.head(top_n),
        dependence_df.head(top_n)
    ]).agent_name
   )
)
top_models
```

Look at overall scores for top models.  
Note: Gemini top models are much better. F1 scores are in range [0.77 - 0.69].

```python
overall_df[overall_df.agent_name.isin(top_models)]
```
